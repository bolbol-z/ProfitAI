```python
# diviser x 
x=dt.iloc[:,:-1]
x
```

```python
# diviser le y = profit 
y=dt.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x["State"]=lb.fit_transform(x["State"])
x
```

```python
# colonnes transformÃ© state => 0,1,2
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough')

x = ct.fit_transform(x)
x
```

```python
x = pd.DataFrame(x, columns=ct.get_feature_names_out())
x
```

```python
x= x.drop(columns=['encoder__State_0'])
x
```

```python
x.head()
```

```python
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
rms = np.sqrt(mean_squared_error(y_test, y_pred))
rms
```

```python
z = rms / y_test.std()
z
```

```python
results = []

for i in range(15, 25): 
    for j in range(0, 5000):
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=i/100, random_state=j)
        
        reg = LinearRegression()
        reg.fit(x_train, y_train)
        y_pred = reg.predict(x_test)
        
        rms = np.sqrt(mean_squared_error(y_test, y_pred))
        z = rms / y_test.std()

            
        

        
        if 0.03 <= z <= 0.06:
            results.append({
                'test_size': i/100,
                'random_state': j,
                'z_score': z
            })


df_results = pd.DataFrame(results)
best_results = df_results.sort_values(by='z_score').head(10)
best_results
```
